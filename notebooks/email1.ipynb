```python
import re
from random import randint
#importing libraries

accountlst = []
x = 1
#building lists and int's

with open("emails.csv", "w") as placehold:
    placehold.write("")
#erasing any old info left in our csv


with open("emails.txt", "r") as file:# opening our email list
    lines = file.readlines() #and reading lines from it
    
    for line in lines:
        compemail = re.findall(r"^(.+)", line)
        compemail = compemail[0]
        #this is helping get rid of any newline characters from line (not necessary but helpful)
        first1 = re.findall(r"^\w", line)
        first1 = first1[0]
        #gets first letter of name
        match = re.search(r"^(\w+)_(\w+)", line)
        if match:
            first, last = match.groups()
        #gets our first and last name (we only need the last)
        matchemail = re.search(r"^(.+)@(\w+)", line)
        if matchemail:
            name, email = matchemail.groups()
        #seperates part of our email and name (we only need part of the email)

        passnum = ""
        for i in range(8):
            num = str(randint(0,9))
            passnum += num
        #builds an 8 digit number for our password

        accountname = first1.upper() + last.lower() + "-" + email
        #builds our account name with previous info

        while accountname in accountlst:
            x += 1
            accountname = accountname + str(x)
        else:
            x = 1
            accountlst.append(accountname)
        #this will look for duplicate accounts and chnge names accordingly
    
        password = first1.upper() + last.lower() + passnum
        #this builds our password with our previous info
        with open("emails.csv", "a") as file2:
            file2.write(compemail + "," + accountname + "," + password + "," + "\n")
        #writes our info to our csv file



```
