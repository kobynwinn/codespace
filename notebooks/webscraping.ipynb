```python
from bs4 import BeautifulSoup
import requests
import re 
#imports our needed libraries for webscraping and regex
page = requests.get("https://www.semrush.com/")
soup = BeautifulSoup(page.content, "html.parser")
#getting our content from the websites
for i in soup.findAll(rel="preconnect"): #finds our links and uses regex to make it more readable
    j = i.get("href")
    j = re.search(r"^https:\/\/\.?(.+)\/?$", j)
    if j:
        j = j.groups()
        print(j[0])
```

```python
from bs4 import BeautifulSoup
import requests
#importing needed libraries for webscraping
page = requests.get("https://www.simplyrecipes.com/how-to-make-a-burger-on-the-stove-6890850")
soup = BeautifulSoup(page.content, "html.parser")
#determining our webpages as usable elements
ingredients = soup.find_all("div", class_ = "comp structured-ingredients")
steps = soup.find_all("li", class_ = "comp mntl-sc-block-group--LI mntl-sc-block mntl-sc-block-startgroup")
title = soup.find("h2", id = "recipe-block__header_1-0")
totaltime = soup.find_all("span", class_ = "comp meta-text")
author = soup.find("a", class_ = "mntl-attribution__item-name").text
# finding all our needed data for the recipe
creditkiller = "Simply Recipes / " + author
#this will build a string that will help us delete any images that got converted to text

print(title.text) #title of our recipe

for i in totaltime: #this will print out the total time of the recipe as well as serving size as a bonus
    text = i.text.replace("\n", " ")
    print(text)


print("Ingredients:") #this will print our ingredients list
for i in ingredients:
    text = i.text.replace("\n\n", "")
    print(text + "\n")


counter = 0 #this will be a counter to number our steps
for i in range(len(steps)): #this prints our steps for our recipe numbered
    counter += 1
    text = steps[i].text.replace("\n\n\n", "").replace(".", ".\n").replace(creditkiller, "")
    print("Step",str(counter) + ":", text)


```
